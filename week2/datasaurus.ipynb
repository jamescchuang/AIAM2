```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ori = pd.read_csv("data/DatasaurusDozen.tsv", delimiter='\t')
```

```python
all_datasets = df_ori.loc[:, 'dataset'].unique()
all_datasets
```

```python
for i in all_datasets:
    df = df_ori[df_ori.loc[:, 'dataset'].isin([i])]
    
    x_mean = 'X Mean: %.4f' % df['x'].mean()
    y_mean = 'Y Mean: %.4f' % df['y'].mean()
    x_std = 'X Std: %.4f' % df['x'].std()
    y_std = 'Y Std: %.4f' % df['y'].std()
    corrxy = 'Corr: %.4f' % df['x'].corr(df['y'])

    font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
        
    plt.scatter(df['x'], df['y'])
    plt.title(i)
    plt.xlim(xmax=100)
    plt.ylim(ymax=110)
    plt.text(105, 50, x_mean, fontdict=font)
    plt.text(105, 40, y_mean, fontdict=font)
    plt.text(105, 30, x_std, fontdict=font)
    plt.text(105, 20, y_std, fontdict=font)
    plt.text(105, 10, corrxy, fontdict=font)
    plt.show()
```
